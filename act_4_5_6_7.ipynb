4. Derivar componentes temporales útiles para el análisis, como año, mes o día de la semana, cuando los datos lo permitan.


```python
#Realizamos importación de libreria datetime
from datetime import date

hoy = pd.Timestamp.today()

# ── Demographics ───
# Año de ingreso a Sigma (útil para análisis de cohortes)
df_demographics['entry_sigma_year'] = df_demographics['date_of_entry_sigma'].dt.year

# Mes de ingreso (¿hay estacionalidad en contrataciones?)
df_demographics['entry_sigma_month'] = df_demographics['date_of_entry_sigma'].dt.month

# Antigüedad recalculada desde fecha real (más precisa que el campo original)
df_demographics['tenure_sigma_calculated'] = (
    (hoy - df_demographics['date_of_entry_sigma']).dt.days / 365.25
).round(2)

# Edad recalculada desde fecha real
df_demographics['age_calculated'] = (
    (hoy - df_demographics['date_of_birth']).dt.days / 365.25
).astype(int)

# ── Succession ───
# Año y mes de nominación
df_succession['nomination_year']  = df_succession['nomination_date'].dt.year
df_succession['nomination_month'] = df_succession['nomination_date'].dt.month

# Edad del incumbente calculada a.k.a. quién ostenta el la vacante a ser sedida por un candidato potencial
df_succession['incumbent_age_calculated'] = (
    (hoy - df_succession['incumbent_dob']).dt.days / 365.25
).astype(int)

print('Ejemplos de "Componentes temporales derivados":\n')
print('Demographics:')
print(df_demographics[['date_of_entry_sigma','entry_sigma_year',
                        'entry_sigma_month','tenure_sigma_calculated',
                        'age_calculated']].head(3).to_string())
print('\nSuccession:')
print(df_succession[['nomination_date','nomination_year',
                     'nomination_month','incumbent_age_calculated']].head(3).to_string())
```

5. Crear una columna calculada relevante mediante una operación entre columnas existentes.

```python
# Columna calculada a partir de los años que le faltan al incumbente para llegar a 65 años
# (edad típica de retiro ejecutivo — indica urgencia de sucesión)
df_succession['years_to_retirement'] = (
    65 - df_succession['incumbent_age_calculated']
).clip(lower=0)  # evita valores negativos si ya pasó los 65

# Clasificar urgencia de sucesión
df_succession['succession_urgency'] = pd.cut(
    df_succession['years_to_retirement'],
    bins=[-1, 3, 7, 15, 100],
    labels=['Crítica', 'Alta', 'Media', 'Baja']
)

print('Columna calculada: urgencia de sucesión\n')
print(df_succession[['incumbent_id', 'incumbent_age_calculated',
                      'years_to_retirement',
                      'succession_urgency']].sample(10).to_string())

print('\nDistribución de urgencia:')
print(df_succession['succession_urgency'].value_counts())
```

```python
# Se define una función para formatear valores numéricos según su unidad de medida
def fmt(val, unit, decimals=2):
    """
    Formatea un valor numérico según su unidad de medida.
    """
    # Si el valor es None, se retorna "N/A"
    if val is None:
        return "N/A"
    # Si la unidad es "MXN", se formatea como moneda mexicana
    if unit == "MXN":
        return f"${val:,.0f} MXN"
    # Si la unidad es "USD", se formatea como moneda estadounidense
    if unit == "USD":
        return f"${val:,.0f} USD"
    # Si la unidad es "%", se formatea como porcentaje
    if unit == "%":
        return f"{val:.{decimals}f}%"
    # Por defecto, se formatea con el número de decimales especificado
    return f"{val:.{decimals}f} {unit}".strip()
```

```python
def statistical_analysis(series: pd.Series, name: str, meta: dict) -> dict:
    """
    Calcula medidas de dispersión para una serie de datos númericos

    Retorna diccionario con:
    a) Medidas de tendencia central: media, mediana, moda
    b) Medidas de dispersión: std, varianza, rango, IQR, CV
    c) Cuartiles: Q1, Q2, Q3
    d) Coeficiente de variación (CV) - solo si la media es mayor de cero y sin valores negativos
    e) Cercas IQR: lower_fence, upper_fence
    f) Valores atípicos: outliers
    """
    # Eliminamos valores nulos
    s = series.dropna().astype(float)

    # Se obtiene el tamaño de la muestra
    n = len(s)

    """ a) Medidas de tendencia central: media, mediana, moda """
    # Media
    mean = s.mean()

    # Mediana
    median = s.median()
    
    # Moda
    # Se obtiene la moda, si existe; de lo contrario, se asigna NaN
    mode = s.mode().iloc[0] if not s.mode().empty else np.nan

    # Medida más representativa y simetría de la distribución
    relative_difference = abs(mean - median) / abs(mean) if mean != 0 else 0

    # Se evalua la diferencia relativa
    if relative_difference <= 0.05:
        representative_measure = "Media. La media y mediana son muy similares"
        simetric_inference = "Se infiere una distribución aproximadamente simetrica"
    elif mean > median > mode:
        representative_measure = "Mediana. Se observa un patrón media > mediana > moda"
        simetric_inference = "Se infiere un posible sesgo positivo"
    elif mean < median < mode:
        representative_measure = "Mediana. Se observa un patrón media < mediana < moda"
        simetric_inference = "Se infiere un posible sesgo negativo"
    else:
        representative_measure = "No hay una medida claramente representativa"
        simetric_inference = "Se requiere revisión manual"

    # Variables donde la media y la mediana difieren de manera notable
    # y explicar qué indica esa diferencia.

    # Diferencia entre media y mediana y porcentaje
    difference = abs(mean - median)
    difference_pct = difference / abs(mean) * 100 if mean != 0 else 0

    # Se evalua el porcentaje de la diferencia
    if difference_pct < 5:
        difference_explanation = "No hay diferencia notable entre la media y mediana, indicando una distribución relativamente simétrica."
    elif mean > median:
        difference_explanation = "La media es mayor que la mediana, lo que sugiere una asimetría positiva (cola hacia la derecha)."
    else:
        difference_explanation = "La media es menor que la mediana, lo que indica una posible asimetría negativa (cola hacia la izquierda)."

    """ ACT 7 b) Medidas de dispersión: std, varianza, rango, IQR y CV """
    
    # Desviación estándar
    std = s.std(ddof=1)

    #Varianza
    variance = s.var(ddof=1)

    # Rango
    rng = s.max() - s.min()

    # Coeficiente de variación
    # Se interpreta cuánto varía la desviación estándar respecto de su valor central (media)
    cv = std / mean * 100 if (meta["cv_valid"] and mean > 0) else None

    # Clasificación del coeficiente de variación
    if cv is None:
        cv_classification = "N/A"
        cv_explanation = "No se puede calcular el CV debido a que la media es cero o hay valores negativos."
    if cv < 20:
        cv_classification = "Muy baja (<20%)"
        cv_explanation = "La variación es muy baja, lo que indica una distribución relativamente homogénea."
    elif cv < 35:
        cv_classification = "Baja (20–35%)"
        cv_explanation = "La variación es baja, lo que indica una distribución relativamente homogénea."
    elif cv < 50:
        cv_classification = "Media (35–50%)"
        cv_explanation = "La variación es media, lo que indica una distribución con una variabilidad moderada."
    else:
        cv_classification = "Alta (>50%)"
        cv_explanation = "La variación es alta, lo que indica una distribución con una gran dispersión de valores."

    """ c) Cuartiles """
    # Q1 (Cuarto 1)
    q1 = s.quantile(0.25)

    # Q2 (Cuarto 2) igual a la mediana
    q2 = s.quantile(0.50)

    # Q3 (Cuarto 3)
    q3 = s.quantile(0.75)

    # Rango intercuartílico
    iqr = q3 - q1  

    # Percentil 10
    p10 = s.quantile(0.10)

    # Percentil 90
    p90 = s.quantile(0.90)


    """e) Cercas IQR """
    # Cerca inferior
    lower_fence = q1 - 1.5 * iqr

    # Cerca superior
    upper_fence = q3 + 1.5 * iqr

    #Cerca inferior extrema
    lower_fence_extreme = q1 - 3.0 * iqr
    
    # Cerca superior extrema
    upper_fence_extreme = q3 + 3.0 * iqr

    """f) Valores atípicos """
    # Valores atípicos por debajo de la cerca inferior
    outliers_low = s[s < lower_fence]

    # Valores atípicos por encima de la cerca superior
    outliers_high = s[s > upper_fence]

    # Valores atípicos por encima de la cerca superior extrema
    extreme_low = s[s < lower_fence_extreme]

    # Valores atípicos por debajo de la cerca inferior extrema
    extreme_high = s[s > upper_fence_extreme]

    # Número total de valores atípicos
    n_out = len(outliers_low) + len(outliers_high)

    # Número total de valores atípicos extremos
    n_ext = len(extreme_low) + len(extreme_high)

    # Clasificación de la cantidad de valores atípicos
    if n_out == 0:
        outlier_classification = "No hay valores atípicos"
    elif n_out <= 5:
        outlier_classification = "Pocos valores atípicos (1–5)"
    elif n_out <= 10:
        outlier_classification = "Algunos valores atípicos (6–10)"
    else:
        outlier_classification = "Muchos valores atípicos (>10)"

    # Porcentaje de valores atípicos
    pct_out = (n_out / n * 100) if n > 0 else 0

    # Cálculo de la asimetría (skewness)
    skew = s.skew()

    # Cálculo de la curtosis (kurtosis)
    kurt = s.kurtosis()

    # Se retorna un diccionario con los resultados del análisis
    return{
        "name"  : name,
        "label" : meta["label"],
        "unit"  : meta["unit"],
        "source": meta["source"],
        "n"     : n,
        "mean"  : mean,
        "median": median,
        "mode"  : mode,
        "representative_measure": representative_measure,
        "simetric_inference": simetric_inference,
        "difference_pct": difference_pct,
        "difference_explanation": difference_explanation,
        "std"   : std,
        "variance": variance,
        "range" : rng,
        "q1"    : q1,
        "q2"    : q2,
        "q3"    : q3,
        "iqr"   : iqr,
        "p10"   : p10,
        "p90"   : p90,
        "cv"    : cv,
        "cv_classification": cv_classification,
        "cv_explanation": cv_explanation,
        "lower_fence": lower_fence,
        "upper_fence": upper_fence,
        "lower_fence_extreme": lower_fence_extreme,
        "upper_fence_extreme": upper_fence_extreme,
        "outliers_low": outliers_low,
        "outliers_high": outliers_high,
        "extreme_low": extreme_low,
        "extreme_high": extreme_high,
        "n_out" : n_out,
        "n_ext" : n_ext,
        "pct_out": pct_out,
        "skew"  : skew,
        "kurt"  : kurt,
    }
```

```python
# Se realiza el bucle para analizar cada variable definida en el diccionario VARIABLES 
# y almacenar los resultados en un nuevo diccionario results
results = {}

for var_name, meta in VARIABLES.items():
    results[var_name] = statistical_analysis(meta["series"], var_name, meta)
```
